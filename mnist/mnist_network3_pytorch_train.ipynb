```python
import mnist_loader
import mnist_network3_pytorch

training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
```

```python
net = mnist_network3_pytorch.Network([
                mnist_network3_pytorch.ConvPoolLayer(1, 20, 5, 2),
                mnist_network3_pytorch.FullyConnectedLayer(20*12*12, 100, dropout_rate=0.5),
                mnist_network3_pytorch.SoftmaxLayer(100, 10)
            ])
net.train_network(training_data, 10, 10, 0.1, test_data=test_data)
```
